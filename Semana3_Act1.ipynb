```python
input = 'https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv'
```

```python
import pandas as pd
dataframe = pd.read_csv(input) #SE LEEN LOS DATOS CSV Y SE ALMACENAN EN LA VARIABLE DE dataframe
dataframe.head()
```

```python
dataframe.dropna(thresh=21, inplace = True) #Este código elimina las filas que tengan dos o más valores nulos
dataframe
```

```python
dataframe['X5'].fillna(value = dataframe.X5.median(),inplace = True) #Se remplaza los valores de la columna X5 con la mediana
```

```python
dataframe['X8'].fillna(value = dataframe.X8.mean(),inplace = True) #Se remplaza los valores de la columnas con la media
dataframe['X9'].fillna(value = dataframe.X9.mean(),inplace = True)
dataframe['X10'].fillna(value = dataframe.X10.mean(),inplace = True)
dataframe['X11'].fillna(value = dataframe.X11.mean(),inplace = True)
dataframe['X14'].fillna(value = dataframe.X14.mean(),inplace = True)
dataframe['X15'].fillna(value = dataframe.X15.mean(),inplace = True)
dataframe['X16'].fillna(value = dataframe.X16.mean(),inplace = True)
dataframe['X17'].fillna(value = dataframe.X17.mean(),inplace = True)
dataframe['X19'].fillna(value = dataframe.X19.mean(),inplace = True)
dataframe['X21'].fillna(value = dataframe.X21.mean(),inplace = True)
dataframe['X22'].fillna(value = dataframe.X22.mean(),inplace = True)
dataframe['X23'].fillna(value = dataframe.X23.mean(),inplace = True)
```

```python
dataframe.info() #SE OBTIENE LA INFO DEL DATAFRAME
```

```python
dfa = dataframe.copy() #SE CREA UN NUEVO DATAFRAME DE ANÁLISIS Y SE ELIMINAN LAS VARIABLES CATEGÓRICAS
dfa.drop('ID',inplace=True, axis=1)
dfa.drop('X2',inplace=True, axis=1)
dfa.drop('X3',inplace=True, axis=1)
dfa.drop('X4',inplace=True, axis=1)
dfa.drop('X6',inplace=True, axis=1)
dfa.drop('X7',inplace=True, axis=1)
dfa.drop('X8',inplace=True, axis=1)
dfa.drop('X9',inplace=True, axis=1)
dfa.drop('X10',inplace=True, axis=1)
dfa.drop('X11',inplace=True, axis=1)
dfa
```

```python
import seaborn as sns
corrs = dfa.corr()
sns.set(rc = {'figure.figsize':(10,5)})
sns.heatmap(corrs, vmin = -1, vmax = 1, cmap = "BuGn", annot= True) #SE CREA LA MATRIZ DE CORRELACIÓN 
```

Como se puede observar, las columnas que más correlación tienen son las variables de X12 a X17.

```python
columnas = ["X12","X13","X14","X15","X16","X17"]
resumen = {
"Varianzas": dfa[columnas].var()[0:6],
"Valor Min": dfa[columnas].min()[0:6],
"Valor Max": dfa[columnas].max()[0:6]
}
pd.DataFrame(resumen, index = columnas).transpose() #SE OBTIENEN LOS VALORES DE LA VARIANZA, EL VALOR MÍNIMO Y EL MÁXIMO
```

A simple vista no nos dice nada, no sabemos bien cuál es mejor o cuál es peor; por lo tanto, se hace la normalización de los datos.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(dfa)
scaled = pd.DataFrame(scaled, columns=dfa.columns)
scaled.head()
```

```python
resumen2 = {
"Varianzas": scaled[columnas].var()[0:6],
"Valor Min": scaled[columnas].min()[0:6],
"Valor Max": scaled[columnas].max()[0:6]
}
pd.DataFrame(resumen2, index = columnas).transpose()
```

```python
from sklearn.decomposition import PCA
pcs = PCA()
pcs_t = pcs.fit_transform(scaled)
```

```python
import numpy as np
pcsResumen = pd.DataFrame({
'% varianza explicada': np.round(pcs.explained_variance_ratio_,4) * 100,
'% varianza acumulada': np.cumsum(pcs.explained_variance_ratio_) * 100
})
pcsResumen #A continuación se muestran las varianzas de las variables
```

```python
pcs_labels = [f'PC{i + 1}' for i in range(len(scaled.columns))]
pcsResumen.index = pcs_labels
pcsResumen #Se consiguen los PCA y su impacto con la información que tienen
```

```python
pcs_df = pd.DataFrame(pcs_t, columns =pcs_labels)
total_var =scaled.var().sum()
pd.DataFrame({
"Porcentaje Varianza": (scaled.var()/ total_var) * 100,
"Porcentaje Varianza Acumulado": (scaled.var().cumsum() / total_var) *100
})

comps_df = pd.DataFrame(
pcs.components_.round(4), 
columns = pcs_df.columns, 
index = scaled.columns) 

comps_df.iloc[:,:8] #Se ve el impacto de cada variable sobre cada PCA
```

```python
comps_df.iloc[:,:15].abs().idxmax() #Se obtiene la variable con mayor impacto en cada PCA
```

Se recomienda usar un PCA cuando se tienen muchas variables. Es un poco tardado trabajar con grandes cantidades de variables y al final puede que se obtenga un resultado similar con menos esfuerzo.
